```python
# stack implementation using singly linkedlist

class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
        
class Stack:
    def __init__(self):
        self.head = None
    
    
    def push(self, value):
        if self.head == None:
            self.head = Node(value)
        else:
            new_node = Node(value)
            new_node.next = self.head
            self.head = new_node
        print("Element added successfully.\n")
    
    def pop(self):
        if self.head == None:
            return "Underflow! Stack is empty."
        else:
            popped_element = self.head.data
            self.head = self.head.next
            print("Element poped successfully.")
            return popped_element
    
s = Stack()

print("Stack operarions:-\n1.Push operation\n2.Pop operation\n3.Exit")

while True:
    choice = int(input("Enter your choice: "))
    if choice == 1:
        element = int(input("Enter element's value: "))
        s.push(element)
    elif choice == 2:
        val = s.pop()
        print(val, end="\n\n")
    elif choice == 3:
        break
        
```

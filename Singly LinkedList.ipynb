```python
# Program for Singly LinkedList

# operations on Singly LL
# 1.Inserting an element
#   -inserting at beginning
#   -inserting at last
#   -inserting at specified location
# 2.Deletion of element
#   -deletion at beginning
#   -deletion at last
#   -deletion at specified location
# 3.Searching an element
# 4.Deleting the entire list
```

```python
class Node:
    def __init__(self, value):
        self.info = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        self.last = None
        
    def inStart(self, value):
        if self.head == None:
            self.head = Node(value)
            self.last = self.head
        else:
            newNode = Node(value)
            newNode.next = self.head
            self.head = newNode
        print("Element added successfully(beginning)")
        
    def inLoc(self, value, loc):
        if self.head == None:
            print("Underflow! LinkedList is empty.")
        else:
            ptr = self.head
            c = 1
            
            while (c != loc) and (ptr.next != None):
                c = c+1
                ptr = ptr.next
                
            if (ptr.next == None):
                print("Specified node is not found.")
            else:
                newNode = Node(value)
                newNode.next = ptr.next
                ptr.next = newNode
        print("Element added successfully(",loc,"index).")
    
    def inLast(self, value):
        if self.head == None:
            self.head = Node(value)
            self.last = self.head
        else:
            newNode = Node(value)
            self.last.next = newNode
            self.last = newNode
        print("Element added successfully(last).")
        
    def outStart(self):
        if self.head == None:
            print("Underflow! LinkedList is empty.")
        else:
            print("Deleted element(at beginning)", self.head.info)
            self.head = self.head.next
    
    def outLoc(self, loc):
        if self.head == None:
            print("Underflow! LinkedList is empty.")
        else:
            ptr1 = self.head
            c = 1
            
            
            while (c != loc) and (ptr1.next != None):
                c = c+1
                ptr1 = ptr1.next
                
            if (ptr1.next == None):
                print("Specified node is not found.")
            else:
                ptr2 = ptr1.next
                print("Deleted element(at index)", ptr2.info)
                ptr1.next = ptr2.next
                
    def outLast(self):
        if self.head == None:
            print("Underflow! LinkedList is empty.")
        else:
            ptr = self.head
            while ptr.next != None:
                l = ptr
                ptr = ptr.next

            print("Deleted element(at last)", ptr.info)
            self.last = l
            self.last.next = None
    
    def search(self, value):
        if self.head == None:
            print("Underflow! List is empty.")
        else:
            c = 0
            ptr = self.head
            while ptr.next != None:
                if ptr.info == value:
                    print("Element found at",c,"index")
                    break
                else:
                    ptr = ptr.next
                    c = c+1
                
    def delete(self):
        print("Deleted list: ", end=" ")        
        while(self.head != None):
            print(self.head.info, end = " ")
            self.head = self.head.next
                        
    def show(self):
        if self.head != None:
            print("Elements in LinkedList: ",end="")
            iter = self.head
            while True:
                print(iter.info, end=" ")
                if iter.next == None:
                    break
                else:
                    iter = iter.next
            print("\n")
        else:
            print("LinkedList is empty.")
            
ll = LinkedList()

print("LinkedList operarions:-\n1.Inserting an element at beginning\n2.Inserting an element at last\n3.Inserting an element at specified location\n4.Deletion at beginning\n5.Deletion at last\n6.Deletion at specified location\n7.Searching an element\n8.Deleting the entire list\n9.Delete entire list\n10.Exit")

while True:
    choice = int(input("Enter your choice: "))
    if choice == 1:
        element = int(input("Enter element's value: "))
        ll.inStart(element)
    elif choice == 2:
        element = int(input("Enter element's value: "))
        Location = int(input("Enter element's Location: "))
        ll.inLoc(element,Location)
    elif choice == 3:
        element = int(input("Enter element's value: "))
        ll.inLast(element)
    elif choice == 4:
        ll.outStart()
    elif choice == 5:
        Location = int(input("Enter element's Location: "))
        ll.outLoc(Location)
    elif choice == 6:
        ll.outLast()
    elif choice == 7:
        element = int(input("Enter element's value: "))
        ll.search(element)
    elif choice == 8:
        ll.delete()
    elif choice == 9:
        ll.show()
    elif choice == 10:
        break
```
